```python
import pandas as pd
```

```python
# List to store each year's DataFrame
dfs = []

# Loop through years 1970 to 2021
for year in range(1970, 2022):
    url = f"https://raw.githubusercontent.com/fantasydatapros/data/master/yearly/{year}.csv"
    try:
        df = pd.read_csv(url)
        df['Year'] = year  # Add the year as a new column
        dfs.append(df)
        print(f"✅ Loaded {year}")
    except Exception as e:
        print(f"⚠️ Could not load {year}: {e}")

# Concatenate all DataFrames into one
full_df = pd.concat(dfs, ignore_index=True)

# Check result
print(f"\nFinal dataset contains {len(full_df)} rows and {len(full_df.columns)} columns.")
```

```python
full_df
```

```python
full_df = full_df[['Player', 'Tm', 'Pos', 'FantasyPoints']]
```

```python
full_df
```

```python
# Ensure FantasyPoints is numeric
full_df['FantasyPoints'] = pd.to_numeric(full_df['FantasyPoints'], errors='coerce')

# Drop rows with missing key data
full_df = full_df.dropna(subset=['Player', 'FantasyPoints', 'Tm'])

# Sort so best seasons are at the top per player/team
full_df_sorted = full_df.sort_values(by='FantasyPoints', ascending=False)

# Group by Player and team, keep the best (first) row for each
best_seasons_df = full_df_sorted.groupby(['Player', 'Tm'], as_index=False).first()

# Optional: Reset index
best_seasons_df = best_seasons_df.reset_index(drop=True)
```

```python
#get all rows where Tm starts with a number and sort by highest fantasy points
best_seasons_df[best_seasons_df['Tm'].str.match(r'^\d')].sort_values(by='FantasyPoints', ascending=False)
```

```python
#drop rows where Tm starts with a number
best_seasons_df.drop(best_seasons_df[best_seasons_df['Tm'].str.match(r'^\d')].index, inplace=True)
```

```python
best_seasons_df[best_seasons_df['Tm'].str.match(r'^\d')].sort_values(by='FantasyPoints', ascending=False)
```
